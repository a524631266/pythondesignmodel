```python
# 作用域代表着一个变量声明周期的状态
# 在python中，查找，改变或者创建变量（crud）都是在一个namespace（命名空间）
# 这个命名空间就是一个存放变量的一块地址（磁盘或者内存，一般是在内存中）
# 我们用sys.modules可以查看内存空间中的变量
# 作用域等价于（数学术语）命名空间，角度不一样，指的是同样一个东西

# 1. 文件作用域命名空间内容
print(dir())
print(locals())
```

```python
# 2. 类命名空间内容
class A:
    name = "zll"
    def p(self):
        print("11")
# 类对象空间
print(A.__dir__(A))
# 对象命名空间
a = A()
print(a.__dir__())
```

```python
# 函数命名空间
class Dog:
    name = "ww"
    def __del__(self):
        print("delete dog")
globall = [1,2,3,4] # global 的append方法，不会导致存储在inner方法本地，global非赋值操作不会存在
def printgloball():
    print(globall)
printgloball()
print("-------")
def a():
    llnotappend = [2,34]
    llneedappend = [3,4] # local 列表的append方法，互导致存储在inner方法本地
    aa = 2
    def inner(): #$$$$$  def 
        b = 2
        print(id(b))
        b += 1
#         global aa # 1.报错，global是对应到文件作用域中
#         aa += 1 # 2.在操作aa的时候没有定义aa，也会报错
        nonlocal aa #$$$$$   3.用nonlocal定义非全局的本地上层作用域
        aa += 1
        print("aa", aa)
        dog = Dog() #$$$$$  赋值
        import sys #$$$$$  import
        globall.append(1) 
        llneedappend.append(3)
        print("inner()",id(b),locals(), id(locals()),id(locals()))
    # 函数每次调用都会在本地创建一个本地作用域（经过测试，是不会变的）,指向新的变量存储地址  
    # 同时会删除变量，看dog就可以了
    inner()
    # 调用相同     
    inner()
    print("dir()",dir())
    print("locals()",locals())
a()
```
